```python
import torch
import os
import numpy as np
from torch.utils.data import DataLoader
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
import text_anomalies as ta
import matplotlib.pyplot as plt
```

# tSne Analysis

Using a pretrained network we perform a tSNE analysis on the latent dim of the network. This script only plots the result see `scripts/tSne_long.py` for how to run.

```python
# Load data
data = ta.DOECDataModule(data_dir="../data/doec/")
data.prepare_data()
data.setup()
dataset = data.dataset
dataloader = DataLoader(
    dataset=dataset,
    batch_size=32,
    num_workers=os.cpu_count(),
    shuffle=False,
    collate_fn=dataset.collate_fn,
)
```

```python
# Load tsne_results (perplexity 500)
tsne_results = np.load(f"../data/doec/tsne_results_{500}.npy")

# Perform DBSCAN to indentify clusters
db = DBSCAN(eps=8, min_samples=50).fit(tsne_results)
labels = db.labels_
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

# Plot
fig, ax =  plt.subplots(figsize=(4, 4))
ax.scatter(tsne_results[:, 0], tsne_results[:, 1], c=labels, cmap="tab20", s=2)

ax.set_xlabel("t-SNE 1 [arb]")
ax.set_ylabel("t-SNE 2 [arb]")

fig.show()

save_kwargs = {
    "bbox_inches": "tight",
    "pad_inches": 0,
    "dpi": 300,
    "transparent": True,
}
fig.savefig("./figures/tsne_analysis.pdf", **save_kwargs)
fig.savefig("./figures/tsne_analysis.png", **save_kwargs)

print("Found {} clusters".format(n_clusters))
```

Savbe assigments to a file.

```python
df = dataloader.dataset.data
df["label"] = labels
# Remove text
df = df.drop(columns=["text"])
```

```python
df.to_csv(f"../data/doec/clusters_{500}.csv", index=False)
```

```python
# Count number of samples in each cluster in percent
for i in range(n_clusters):
    print(
        "Cluster {}: {:.2f}%".format(
            i, 100 * (labels == i).sum() / len(labels)
        )
    )
```
